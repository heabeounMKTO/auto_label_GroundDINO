```python
import torch
from segment_anything import sam_model_registry

DEVICE = torch.device('cpu') #mfw no 3090 at work
MODEL_TYPE = "vit_h"
CHECKPOINT_PATH = "weights/sam_vit_h_4b8939.pth"
sam = sam_model_registry[MODEL_TYPE](checkpoint=CHECKPOINT_PATH)
sam.to(device=DEVICE)
```

```python
import cv2
from segment_anything import SamAutomaticMaskGenerator

img_path = "test_set/t1.jpeg"

mask_gen = SamAutomaticMaskGenerator(sam)
imgae = cv2.cvtColor((cv2.imread(img_path)), cv2.COLOR_BGR2RGB)
result = mask_gen.generate(imgae)
```

```python
import supervision as sv
mask_annotator = sv.MaskAnnotator()
detections = sv.Detections.from_sam(result)
annotated_image = mask_annotator.annotate(imgae, detections)

%matplotlib inline
sv.plot_image(annotated_image, (16, 16))
```
